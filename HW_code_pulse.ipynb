```python
import matplotlib.pyplot as plt
import numpy as np


def f(x,t, k_0=1.0, c=1.0, sigma=5): 
    
    return 1.0/(2*np.pi)*np.exp(1.0j*k_0*(x-c*t))*np.exp(-(x-c*t)**2*sigma**2/2) 

x = np.linspace(0,2*np.pi, 1000).reshape(1,-1) 
t = np.linspace(-5,5,1000).reshape(-1,1) 

frame_data = f(x, 0, k_0=1/(2*np.pi*6.27*10**-7)).real  #setting this to have a wavelength of 620nm to correspond to red light

figure, ax = plt.subplots(1,1, figsize=(8,4.5)) 
line, = ax.plot(x[0], frame_data[0]) 

ax.set_xlim(0,4)

ax.set_title(r'$f(x,t) = \frac{1}{2\pi} e^{ik_{0}(x-ct)-\frac{(x-ct)^{2}\sigma^{2}}{2}}$', fontsize=20);
ax.set_xlabel(r'$x$', fontsize=12);
ax.set_ylabel(r'$f(x,t)$', fontsize=12);


plt.show()
```
